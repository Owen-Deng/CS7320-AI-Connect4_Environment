# Example of how to use the environment

Define two random agents. Notes:
* You need two different functions so the timing works.
* The agent 

```python
from environment import truly_dynamic_environment, replay, visualize
import numpy as np

def random_agent1(board, player): 
    """A random agent."""
    return np.random.choice(np.where(board[0]==0)[0])
def random_agent2(board, player): 
    """Another random agent."""
    return np.random.choice(np.where(board[0]==0)[0])

players = [
    {
        "algo": random_agent1,
        "player": 1,
        "args": {}
    },
    {
        "algo": random_agent2,
        "player": -1,
        "args": {}
    }
]

```

Run the game

```python

result,final_board,all_boards = truly_dynamic_environment(players, size=(6,7), visual=True)

visualize(final_board)
```

```python
replay(all_boards)
```
